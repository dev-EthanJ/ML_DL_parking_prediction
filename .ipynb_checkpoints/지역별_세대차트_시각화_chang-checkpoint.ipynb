```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 글꼴깨짐 방지
import matplotlib
import matplotlib.font_manager as fm
import seaborn as sns
fm.get_fontconfig_fonts()
font_location = 'C:/Windows/Fonts/malgun.ttf' # For Windows
font_name = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```

```python
df_a = pd.read_csv('C:/Users/82109/머신러닝 딥러닝/ML_DL_parking_prediction/age_gender_info.csv')
df_a.head()
```

```python
df_a.groupby('지역').mean()
```

```python
# age_gender 형태
df_a.shape
```

```python
ag = df_a.set_index('지역')
```

```python
ag.loc['전체 평균'] = ag.mean()
ag.loc['광역시 평균'] = ag.loc[(ag.index.str.contains('시')) & (~ag.index.str.contains('세종'))].mean()
ag.loc['도 평균'] = ag.loc[ag.index.str.contains('도')].mean()
```

```python
fig = plt.figure(figsize = (14, 10))
sns.lineplot(data=ag.T, )
plt.title('지역별 세대 라인 차트')
plt.xticks(rotation=45)
```

```python
fig = plt.figure(figsize = (14, 10))
sns.heatmap(data=ag.T, )
plt.title('지역별 세대 라인 차트')
plt.xticks(rotation=45)
```
